```python
import os
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_directory = os.path.join(os.getcwd(), 'dataset_diabetes')
```

```python
diabetes = pd.read_csv(os.path.join(data_directory, 'diabetic_data.csv'))
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(diabetes, test_size=0.2, random_state=42)
```

```python
train_set.to_csv(os.path.join(data_directory, 'diabetic_data_train.csv'), index=False)
```

```python
test_set.to_csv(os.path.join(data_directory, 'diabetic_data_test.csv'), index=False)
```
